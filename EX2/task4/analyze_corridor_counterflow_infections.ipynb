# ANALYZE INFECTIONS IN CORRIDOR COUNTERFLOW

```python
import pandas as pd
```

Since the scenario's infected pedestrians are automatically generated we already know their ids (100-199) as can be seen in the file create_scenario_task4.5_2.py. We can therefore filter out these pedestrians with the aim of understanding the number of infections only on the susceptible pedestrians.

```python
# read file containing infection information
input_path = "task4.5/output/corridor_sim/SIRInformation.csv"
df = None
with open(input_path) as f:
    df = pd.read_csv(f, sep=" ")
```

```python
# for readability
ID_SUSCEPTIBLE = 1
ID_INFECTIVE = 0
PEDESTRIANS_FROM_SOURCE = 100

# create df containing only susceptible
infective_range = range(100, 200)
sus_df = df[~df['pedestrianId'].isin(infective_range)]

# get initial susceptible count 
init_sus = PEDESTRIANS_FROM_SOURCE

# get final susceptible count
final_sus = init_sus
for ped in sus_df['pedestrianId'].unique():
    if ID_INFECTIVE in sus_df[sus_df['pedestrianId'] == ped]['groupId-PID5'].values:
        final_sus -= 1

print(f"Pedestrians infected traversing the corridor are: {init_sus - final_sus}")
```
