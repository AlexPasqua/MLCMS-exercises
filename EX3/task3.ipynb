**Task3.1**

Plot three phase diagrams of the following system:

$$\dot x_1 = \alpha x_1 - x_2 - x_1(x_1^2 + x_2^2)$$
$$\dot x_2 = x_1 + \alpha x_2 - x_2(x_1^2 + x_2^2)$$

```python
odes = ["alpha * x1 - x2 - x1 * (x1 ** 2 + x2 ** 2)", "x1 + alpha * x2 - x2 * (x1 ** 2 + x2 ** 2)"]
# showing the phase portrait of steady state behaviour
fig = create_phase_portrait_derivative(system=odes, alpha=-1, title_suffix="attracting steady state", fig_size=5)
```

```python
# showing the phase portrait at bifurcation point
create_phase_portrait_derivative(system=odes, alpha=0, title_suffix="bifurcation point", fig_size=5)
```

```python
# showing the phase portrait of limit cycle behaviour
create_phase_portrait_derivative(system=odes, alpha=1, title_suffix="attracting limit cycle", fig_size=5)
```
