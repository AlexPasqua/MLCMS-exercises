```python
import numpy as np
from utilities import create_phase_portrait
```

Create the phase portraits shown in figure 2.5 at page 49 of the book
_Elements of Applied Bifurcation Theory_ by Kuznetsov

**Unstable focus:**

```python
# unstable focus
alpha = 0.1
A = np.array([
    [alpha, alpha],
    [-1/4, 0]
])
create_phase_portrait(A, alpha, title_suffix="unstable focus")
```

**Unstable saddle:**

```python
# unstable saddle
alpha = -0.2
A = np.array([
    [-alpha, alpha],
    [-1/4, 0]
])
create_phase_portrait(A, alpha, title_suffix="unstable saddle")
```

**Unstable node:**

```python
# unstable node
alpha = 2
A = np.array([
    [alpha, alpha],
    [-1/4, 0]
])
create_phase_portrait(A, alpha, title_suffix="unstable node")
```

**Stable node:**

```python
# stable node
alpha = -0.25
A = np.array([
    [alpha, 0],
    [0, -1]
])
create_phase_portrait(A, alpha, title_suffix="node stable")
```

**Stable focus:**

```python
# stable focus
alpha = 0.1
A = np.array([
    [-alpha, alpha],
    [-1/4, 0]
])
create_phase_portrait(A, alpha, title_suffix="stable focus")
```
